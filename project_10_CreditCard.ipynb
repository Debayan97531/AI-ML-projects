```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
credit_card=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card.head()
```

```python
credit_card.tail()
```

```python
credit_card.info()
```

```python
credit_card.isnull().sum()
```

```python
#distribution of legit and fraud data
credit_card['Class'].value_counts()
```

Dataset is highly unbalanced

```python
#separating data for analysis
legit = credit_card[credit_card.Class==0]
fraud = credit_card[credit_card.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures for data
#describing amount column
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transaction
credit_card.groupby('Class').mean()
```

UnderSampling

Build a dataset containing similar distributionn of normal transactions and fraud transactions

```python
legit_sample=legit.sample(n=492)
```

```python
#concatenating two datasets
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
train_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on train data',train_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_prediction=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:',test_data_prediction)
```
